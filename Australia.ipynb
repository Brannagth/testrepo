El objetivo de este gráfico es visualizar la relación entre la cantidad de lluvia caída y el riesgo asociado en diferentes años. La cantidad de lluvia y el riesgo son variables importantes a considerar para comprender el impacto de las condiciones climáticas en un área determinada.

Al representar los datos en un gráfico de barras, se busca comparar visualmente el promedio de la cantidad de lluvia caída y el promedio del riesgo asociado en cada año. Esta representación permite identificar posibles tendencias o patrones a lo largo del tiempo y comprender cómo han variado estos aspectos en diferentes años.

Al observar el gráfico, se puede analizar si existe alguna relación entre la cantidad de lluvia y el riesgo, es decir, si a mayor cantidad de lluvia caída se asocia un mayor riesgo. También se puede identificar si ciertos años han presentado un mayor riesgo en comparación con otros, lo cual puede proporcionar información relevante para la planificación y gestión de eventos relacionados con el clima.

```python
data_frame_australia['Date'] = pd.to_datetime(data_frame_australia['Date'])
datos_agrupados = data_frame_australia.groupby(data_frame_australia['Date'].dt.year).agg({'Rainfall': 'mean', 'RISK_MM': 'mean'}).reset_index()
datos_agrupados.plot(x='Date', y=['Rainfall', 'RISK_MM'], kind='bar')

# Configura las etiquetas de los ejes y el título del gráfico
plt.xlabel('Año')
plt.ylabel('Riesgo x cm3 de lluvia')
plt.title('Riesgo V/S cantidad de lluvia caida')
plt.legend(['LLuvia caida (cm3)', 'Riesgo'])
plt.show()
```

###Gráfico de Tendencia

El objetivo de este gráfico es mostrar cómo varía la precipitación mensual a lo largo del tiempo en Australia y proporcionar una visión general de las tendencias y patrones climáticos a lo largo de los meses.

 La precipitación es un factor climático fundamental y su variabilidad tiene un impacto significativo en varios aspectos, como la agricultura, la gestión del agua y la planificación urbana. Al analizar la tendencia de la precipitación mensual, podemos obtener información valiosa sobre los patrones climáticos estacionales y comprender mejor la distribución de la lluvia a lo largo del año.

  Este gráfico de tendencias de la precipitación mensual cumple el objetivo de visualizar y analizar la variabilidad de la precipitación a lo largo del tiempo. Proporciona información relevante sobre los patrones climáticos estacionales y permite comparar la precipitación mensual con el promedio anual. Además, las etiquetas de valores y el relleno de área brindan información adicional y resaltan la variabilidad mensual de la precipitación.

```python
# Convertir la columna 'Date' a tipo datetime
data_frame_australia['Date'] = pd.to_datetime(data_frame_australia['Date'])

# Crear una columna adicional para el mes
data_frame_australia['Month'] = data_frame_australia['Date'].dt.month

# Calcular la precipitación mensual promedio
monthly_rainfall = data_frame_australia.groupby('Month')['Rainfall'].mean()
annual_rainfall_avg = data_frame_australia['Rainfall'].mean()

# Crear el gráfico de tendencias
plt.figure(figsize=(12, 6))
plt.plot(monthly_rainfall.index, monthly_rainfall.values, color='blue', label='Precipitación mensual promedio')
plt.axhline(y=annual_rainfall_avg, color='red', linestyle='--', label='Promedio anual')

# Etiquetas de valores de precipitación
for x, y in zip(monthly_rainfall.index, monthly_rainfall.values):
    plt.text(x, y, str(round(y, 2)), ha='center', va='bottom')

# Valor de la media
plt.text(1, annual_rainfall_avg, f'Media: {round(annual_rainfall_avg, 2)} mm', ha='left', va='bottom', color='red')

# Relleno de área
plt.fill_between(monthly_rainfall.index, monthly_rainfall.values, alpha=0.3, color='lightblue')

plt.xlabel('Meses')
plt.ylabel('Precipitación Promedio (mm)')
plt.title('Tendencia de la Precipitación Mensual desde 2007 hasta 2017')
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.legend()
plt.tight_layout()
plt.show()
```

Del gráfico de tendencia de la precipitación promedio en Australia durante el período de 2007 a 2017 realizamos las siguientes conclusiones:

*   Se observa que la tendencia general indica que hay una mayor cantidad de lluvia durante los meses de diciembre, enero y febrero, que corresponden al verano en Australia. Estos meses registran valores de precipitación promedio de 2.72 mm, 3.17 mm y 2.48 mm respectivamente. Esta tendencia puede ser atribuida a las frecuentes tormentas tropicales que suelen ocurrir durante el verano en el país.
*   Por otro lado, se observa un segundo pico en el mes de junio, que marca el comienzo del invierno en Australia. Durante este mes, la precipitación promedio alcanza los 2.78 mm. Este aumento en la lluvia puede estar relacionado con las zonas de clima frío presentes en el país durante el invierno.
*   En contraste, los meses de otoño (marzo, abril y mayo) y primavera (septiembre, octubre y noviembre) muestran valores de precipitación por debajo del promedio. Estos meses presentan un clima más templado y, en general, no experimentan una cantidad significativa de lluvia.

En resumen, el análisis del gráfico de tendencia sugiere que la precipitación promedio en Australia sigue un patrón estacional, con mayores niveles de lluvia durante el verano y el invierno, y niveles más bajos durante el otoño y la primavera. La media general de precipitación a lo largo de los diez años analizados es de 2.35 mm.

###Gráfico para visualizar Partes de un total

El propósito principal es comparar la proporción de eventos de lluvia y no lluvia tanto para el día actual como para el día siguiente. El gráfico busca resaltar visualmente las diferencias en las proporciones y proporcionar una representación clara de la distribución de los eventos.

Al mostrar dos círculos de pastel lado a lado, se permite una comparación directa entre las proporciones de eventos de lluvia hoy y mañana. El uso de colores distintos y la separación del segmento 'Llueve' en el segundo círculo de pastel ayuda a destacar la diferencia entre las dos categorías.

Este gráfico busca proporcionar una visualización clara y efectiva de las proporciones de eventos de lluvia hoy y mañana, permitiendo una comparación rápida y fácil entre las dos categorías.

```python
# Gráfico de partes de un total: Comparación de eventos de lluvia hoy y lluvia mañana
plt.figure(figsize=(10, 6))
rain_today = data_frame_australia['RainToday'].value_counts()
rain_tomorrow = data_frame_australia['RainTomorrow'].value_counts()
labels = ['No llueve', 'Llueve']
colors = ['#1f77b4', '#ff7f0e']
explode = (0, 0.1)

plt.subplot(121)
plt.pie(rain_today, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90, shadow=True)
plt.axis('equal')
plt.title('Lluvia Hoy')

plt.subplot(122)
plt.pie(rain_tomorrow, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90, shadow=True)
plt.axis('equal')
plt.title('Lluvia Mañana')

plt.suptitle('Comparación de Eventos de Lluvia Hoy y Mañana')
plt.tight_layout()
plt.show()
```

Al comparar los dos gráficos, se puede observar claramente la similitud en la proporción de los datos. En la variable "RainToday", el 77.7% de los registros corresponden a días sin lluvia, mientras que el 23.3% corresponden a días con lluvia. En la variable "RainTomorrow", se encuentra una proporción muy similar, con un 77.6% de registros indicando que no lloverá mañana y un 22.4% indicando que sí lloverá. La diferencia entre ambas proporciones es mínima, apenas un 0.1%.

Estos resultados nos llevan a concluir que en Australia la mayoría de los días no presenta lluvia. Ambas variables muestran un alto porcentaje de registros indicando ausencia de lluvia, lo que sugiere un clima predominantemente seco en el país. Es notable la consistencia en la proporción de registros entre ambas variables, reforzando la idea de la escasez de eventos de lluvia en general.

En resumen, estos gráficos nos brindan una perspectiva visual y cuantitativa de la falta de lluvia en Australia, respaldando la noción de que el país experimenta en su mayoría condiciones áridas.

###Gráfico de Distribuciones

El gráfico de distribución de las temperaturas mínimas diarias busca representar la variabilidad y la tendencia central de los datos de temperaturas mínimas en un período de tiempo determinado. A través del histograma, se puede observar la distribución de frecuencia de las temperaturas, es decir, cuántas veces se registraron diferentes rangos de temperaturas mínimas. Esto permite identificar patrones, como la concentración de temperaturas en ciertos rangos o la presencia de valores atípicos.

```python
# Gráfico de distribuciones: Temperaturas mínimas diarias
plt.figure(figsize=(8, 6))
sb.histplot(data_frame_australia['MinTemp'], kde=True, color='purple')

# Agregar líneas verticales para representar la media y la mediana
mean_temp = data_frame_australia['MinTemp'].mean()
median_temp = data_frame_australia['MinTemp'].median()
plt.axvline(x=mean_temp, color='red', linestyle='--', label='Media')
plt.axvline(x=median_temp, color='green', linestyle='--', label='Mediana')

# Agregar texto de la media y la mediana
plt.text(mean_temp + 0.2, plt.ylim()[1]*0.9, f'Media: {mean_temp:.2f}', color='red')
plt.text(median_temp + 0.2, plt.ylim()[1]*0.85, f'Mediana: {median_temp:.2f}', color='green')

plt.xlabel('Temperatura Mínima (°C)')
plt.ylabel('Frecuencia')
plt.title('Distribución de las Temperaturas Mínimas Diarias')
plt.legend()
plt.tight_layout()
plt.show()
```

###Gráfico de Correlaciones

El objetivo del gráfico de la matriz de correlación es visualizar las relaciones lineales entre las variables de interés en el dataset. Esta visualización permite identificar la fuerza y la dirección de las correlaciones entre las variables, lo que puede ser útil para entender cómo se relacionan entre sí.

Este gráfico proporciona información valiosa sobre la dependencia lineal entre las variables. Esto puede ayudar a identificar posibles patrones, tendencias o asociaciones entre las variables analizadas.

Se puede identificar qué variables tienen una correlación positiva o negativa más fuerte entre sí. Esto puede ser útil para identificar posibles relaciones directas o inversas entre las variables y puede proporcionar información sobre cómo una variable puede influir en otra.

```python
# Calcular la matriz de correlación
correlation_matrix = data_frame_australia.corr()

# Variables de interés
variables_interes = ['MaxTemp', 'MinTemp', 'Temp9am','Temp3pm','Pressure9am','Pressure3pm','Humidity3pm','Humidity9am','WindSpeed9am','WindSpeed3pm']

# Filtrar la matriz de correlación
correlation_subset = correlation_matrix[variables_interes].loc[variables_interes]

# Crear el gráfico de correlaciones con colores personalizados
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(correlation_subset, dtype=bool))
cmap = sb.color_palette("RdYlBu_r", as_cmap=True)  # Colores personalizados: Rojo-Amarillo-Azul (inverso)
sb.heatmap(correlation_subset, mask=mask, cmap=cmap, annot=True, fmt=".2f", square=True)

# Ajustes estéticos
plt.title('Matriz de Correlación - Variables de Interés')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

plt.show()
```


Al analizar la matriz de correlación entre nuestras variables de interés, se pueden extraer las siguientes conclusiones:

Correlaciones positivas fuertes: Se observan correlaciones positivas fuertes entre las siguientes variables:


  *   MaxTemp y Temp3pm: Estas variables están altamente correlacionadas, con un valor de correlación de 0.98. Esto indica una fuerte relación entre la temperatura máxima del día y la temperatura en la tarde.

  *   MaxTemp y Temp9am: También existe una correlación positiva alta entre la temperatura máxima y la temperatura en la mañana, con un valor de correlación de 0.89.

  *   MinTemp y Temp9am: La temperatura mínima y la temperatura en la mañana presentan una correlación positiva significativa de 0.90.

  *   Temp3pm y Temp9am: Estas variables están altamente correlacionadas, con un coeficiente de correlación de 0.96, lo que indica una fuerte relación entre la temperatura en la tarde y la temperatura en la mañana.



Correlaciones negativas moderadas: Se observan correlaciones negativas moderadas entre las siguientes variables:


  *   Humidity3pm y Temp3pm: Existe una correlación negativa de -0.56 entre la humedad en la tarde y la temperatura en la tarde. Esto sugiere que a medida que la temperatura en la tarde aumenta, la humedad tiende a disminuir.

  *   MaxTemp y Humidity3pm: También se encuentra una correlación negativa moderada de -0.51 entre la temperatura máxima y la humedad en la tarde. Esto indica que a medida que la temperatura máxima aumenta, la humedad tiende a disminuir.

```python
# Reemplazando los valores nulos de la columna RainToday con la moda.
data_frame_australia_Limpia['RainToday'].fillna(data_frame_australia_Limpia['RainToday'].mode()[0], inplace=True)
```

En esta instrucción de código decidimos remplazar los valores nulos de RainToday con la moda ya que al ser una variable categórica es una forma común de tratar valores nulos, ya que la moda representa el valor mas frecuente en la distribución de la variable y podría ser una buena aproximación del valor verdadero. Al incluir estas variables en el modelo de regresión, se busca capturar la variabilidad y los patrones relacionados con la temperatura mínima, lo que permite realizar predicciones más precisas.

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y ajustar el modelo de árbol de decisión de regresión con los hiperparámetros ajustados
model = DecisionTreeRegressor(max_depth=5, min_samples_split=10)
model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Calcular el error cuadrático medio (MSE) y el coeficiente de determinación (R2)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```

El coeficiente de determinación (R2) obtenido es de aproximadamente 0.84. Esta métrica proporciona una medida de qué tan bien se ajustan los valores predichos a los valores reales. #### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Paso 1: Preparar los datos

X = data_frame_australia_Limpia[['Pressure9am', 'WindSpeed9am', 'MaxTemp', 'Humidity9am', 'Temp9am']] # Datos de entrada (características)
y = data_frame_australia_Limpia['MinTemp']  # Variable objetivo (precipitaciones)

# Paso 2: Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Paso 3: Crear el modelo de Bosques Aleatorios
rf = RandomForestRegressor(n_estimators=50, random_state=42)
# n_estimators: número de árboles en el bosque (puedes ajustarlo según tus necesidades)

# Paso 4: Entrenar el modelo
rf.fit(X_train, y_train)

# Paso 5: Realizar predicciones en los datos de prueba
y_pred = rf.predict(X_test)

# Paso 6: Evaluar el rendimiento del modelo
mse = mean_squared_error(y_test, y_pred)
print("Error cuadrático medio (MSE):", mse)

rmse = np.sqrt(mse)
print("Raíz del error cuadrático medio (RMSE):", rmse)

r2 = r2_score(y_test, y_pred)
print("Coeficiente de determinación (R2):", r2)

plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel('Etiquetas reales')
plt.ylabel('Predicciones')
plt.title('Comparación entre etiquetas reales y predicciones')

# Agregar MSE y R2 al gráfico
mse_label = 'MSE: {:.2f}'.format(mse)
r2_label = 'R2 Score: {:.2f}'.format(r2)
plt.text(min(y_test), max(y_pred), mse_label, horizontalalignment='left', verticalalignment='bottom', fontsize=10)
plt.text(min(y_test), max(y_pred), r2_label, horizontalalignment='left', verticalalignment='top', fontsize=10)
plt.show()
```

Los resultados obtenidos indican que el modelo de Random Forest se ajusta a los datos, ya que tanto el MSE como el RMSE son relativamente bajos, acercandose bastante a 0, lo que implica que los valores se encuentran cercanos a los reales. Además, el coeficiente de determinación (R2) indica que el modelo logra reccorer una gran proporción de la variabilidad en los datos, es decir, no puede recorrer la totalidad de los datos entregados pero si la gran mayoria de ellos, abarcando el 88%.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from kneed import KneeLocator

# Seleccionar las variables de interés
data = data_frame_australia_Limpia[['MinTemp', 'MaxTemp']]

# Normalizar los datos
scaler = StandardScaler()
normalized_data = scaler.fit_transform(data)

# Aplicar K-means con diferentes números de clusters
sse = []
for k in range(1, 40):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(normalized_data)
    sse.append(kmeans.inertia_)

# Graficar SSE vs. número de clusters
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(range(1, 40), sse, marker='o')
ax.set_xlabel('Número de Clusters')
ax.set_ylabel('SSE')
ax.set_title('Método del Codo')

# Utilizar el método del codo para determinar el número óptimo de clusters
kl = KneeLocator(range(1, 40), sse, curve='convex', direction='decreasing')
optimal_clusters = kl.elbow
ax.annotate('Codo', xy=(optimal_clusters, sse[optimal_clusters-1]), xytext=(optimal_clusters+3, sse[optimal_clusters-1]+20000),
             arrowprops=dict(arrowstyle='->'))
plt.show()

print('Número óptimo de clusters:', optimal_clusters)
```

La ubicación del "codo" en el gráfico indica el número óptimo de clusters, que en este caso es 6. Esto significa que dividir los datos en 6 clusters puede capturar la estructura y la variabilidad de los datos de manera efectiva, minimizando el SSE.

```python
# Realizar el ajuste del modelo K-Means
kmeans = KMeans(n_clusters=6)
kmeans.fit(data_frame_australia_Limpia[['MinTemp', 'MaxTemp']])

# Obtener las etiquetas de los clusters asignados a cada punto
labels = kmeans.labels_

# Obtener las coordenadas de los centroides
centroids = kmeans.cluster_centers_

# Crear el gráfico de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(data_frame_australia_Limpia['MinTemp'], data_frame_australia_Limpia['MaxTemp'], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroides')
plt.xlabel('Temperatura Mínima')
plt.ylabel('Temperatura Máxima')
plt.title('Gráfico de Clusters')
plt.legend()
plt.show()
```

A partir del gráfico, se puede observar que los datos se han agrupado en 6 clusters distintos, cada uno representado por un color diferente. Estos clusters reflejan patrones o agrupaciones naturales en los datos, donde puntos similares en términos de temperatura mínima y máxima se han asignado al mismo grupo.

Los centroides representan los puntos centrales de cada cluster marcados con una X roja, y su posición indica el promedio de las temperaturas mínimas y máximas de los puntos asignados a ese cluster.
